```python
import os
import csv
import sys
os.getcwd()

os.chdir('C:\\Users\\Leessx\\Desktop\\leezzx\\학교과제\\텍스트마이닝')
f = open('202002_ages.csv')
data = csv.reader(f)
result1 = []
result2 = []

name1 = input('동 이름을 입력하세요: ')
name2 = input('비교할 동 이름을 입력하세요: ')

for row in data :
    if name1 in row[0] :
        for i in row[3:] :
            result1.append(int(i.replace(',', '')))
    elif name2 in row[0] :
        for i in row[3:] :
            result2.append(int(i.replace(',', '')))
        
import matplotlib.pyplot as pl
pl.style.use('ggplot')
pl.rc('font', family = 'Malgun Gothic')
pl.figure(figsize = (15, 9))
pl.title(name1 + '과 ' + name2 + '의 인구구조 비교')
pl.plot(result1, 'r', label = name1)
pl.plot(result2, 'b', label = name2)
pl.legend(loc = 'upper right')
pl.show()
```
