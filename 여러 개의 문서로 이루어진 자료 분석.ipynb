```python
import os
import csv
import math
import pandas as pd

# 함수

# term frequency
def tf(t, d):
    return d.count(t)

# inverse docuent frequency
def idf(t):
    df = 0
    for doc in sdocs:
        df += t in doc
    return math.log(N / (df + 1))

# TF-IDF
def tfidf(t, d):
    return tf(t, d) * idf(t)

os.getcwd()
os.chdir('C:\\Users\\Leessx\\Desktop\\leezzx\\학교과제\텍스트마이닝')

# 데이터 로딩
f = open("삼성전자 주식 종목토론 방 15개 문서.txt", 'r', encoding='UTF8')
docs = f.read()
f.close()

sdocs = docs.split(",")
review_list = list(set(w for doc in sdocs for w in doc.split()))
review_list.sort()

# TF-IDF 계산
N = len(sdocs)

result = []
for i in range(N):
    result.append([])
    d = sdocs[i]
    for j in range(len(review_list)):
        t = review_list[j]
        
        result[-1].append(tfidf(t, d))
        
review_tfidf = pd.DataFrame(result, columns = review_list)
review_tfidf
```

```python
# term frequency 계산
N = len(sdocs)

result = []
for i in range(N):
    result.append([])
    d = sdocs[i]
    for j in range(len(review_list)):
        t = review_list[j]
        
        result[-1].append(tf(t, d))
        
review_tf = pd.DataFrame(result, columns = review_list)
review_tf

# KMeans 군집분석
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 2).fit(review_tf)

cl = model.fit_predict(review_tf)
review_tf['주식'] = cl
review_tf
```
